```python
import random
import itertools
import string

def generate_text_file(filename, words, pattern, file_size, line_num):
    with open(filename, 'w') as f:
        size = 0
        line_size = file_size / line_num
        current_line_size = 0
        
        while size < file_size:
            for word in pattern(words, file_size, size):
                f.write(word)
                f.write(' ')
                size += len(word) + 1
                current_line_size += len(word) + 1

                # Check if we've reached the desired number of characters per line
                if current_line_size >= line_size and size < file_size:
                    f.write('\n')
                    current_line_size = 0


current_word_index = 0

def pattern_alternating(words, file_size, size):
    global current_word_index
    word = words[current_word_index]
    current_word_index = (current_word_index + 1) % len(words)
    return [word]

def pattern_batches(words, file_size, size):
    wordA, wordB, wordC, wordD = words
    quarter_file_size = file_size // 4

    if size < quarter_file_size:
        return [wordA]
    elif size < 2 * quarter_file_size:
        return [wordB]
    elif size < 3 * quarter_file_size:
        return [wordC]
    else:
        return [wordD]

def pattern_random_uniform(words, file_size, size):
    word_length_sum = sum(len(word) for word in words)
    total_words = file_size // (word_length_sum + len(words))
    uniform_list = [word for word in words for _ in range(total_words // len(words))]

    random.shuffle(uniform_list)
    uniform_cycle = itertools.cycle(uniform_list)

    while size < file_size:
        word = next(uniform_cycle)
        yield word
        size += len(word) + 1

def random_word(length):
    return ''.join(random.choice(string.ascii_letters) for _ in range(length))

def generate_random_words_file(filename, num_words, word_length, file_size, line_num):
    words = [random_word(word_length) for _ in range(num_words)]
    with open(filename, 'w') as f:
        size = 0
        line_size = file_size / line_num
        current_line_size = 0

        while size < file_size:
            word = random.choice(words)
            f.write(word)
            f.write(' ')
            size += len(word) + 1
            current_line_size += len(word) + 1

            # Check if we've reached the desired number of characters per line
            if current_line_size >= line_size and size < file_size:
                f.write('\n')
                current_line_size = 0

word_sets = [
    ('sugar', 'dream', 'piano', 'spicy'),
    # Add more 4-word sets here
]

file_size = 50 * 1024 * 1024  # 50 MB
line_num = 500

for y in range(999,1001):
    for i, words_set in enumerate(word_sets):
         generate_text_file(f'textfile_{i}_pattern_alternating_lines_{y}.txt', words_set, pattern_alternating, file_size, y)
         generate_text_file(f'textfile_{i}_pattern_batches_lines_{y}.txt', words_set, pattern_batches, file_size, y)
         generate_text_file(f'textfile_{i}_pattern_random_uniform_lines_{y}.txt', words_set, pattern_random_uniform, file_size, y)



#for y in range(1,4):
#    generate_random_words_file(f'random_words_lines_{y}.txt', 16, 5, file_size, y)

```
